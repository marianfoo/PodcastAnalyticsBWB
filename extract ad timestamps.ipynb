```python
import numpy as np
from scipy.io import wavfile
from scipy import signal
import os
import csv   
import time
```

```python
t = time.process_time()
for i, filename in enumerate(os.listdir(r"E:\files\spotify_wav")):
    ad_start = 'ad_start.wav'
    ad_end = 'ad_end.wav'
    filename = os.listdir(r"E:\files\spotify_wav")[i]
    source = r'E:\files\spotify_wav\{}'.format(filename)
    print(source)
    # read the sample to look for
    rate_ad_start, ad_start = wavfile.read(ad_start);
    ad_start = np.array(ad_start[:,0], dtype='float')
    rate_ad_end, ad_end = wavfile.read(ad_end);
    ad_end = np.array(ad_end[:,0], dtype='float')
    ## read the source
    rate, source = wavfile.read(source);
    source = np.array(source[:,0], dtype='float')
    # resample such that both signals are at the same sampling rate (if required)
    if rate != rate_ad_start:
        num = int(np.round(rate*len(ad_start)/rate_ad_start))
        ad_start = signal.resample(ad_start, num)
    if rate != rate_ad_end:
        num = int(np.round(rate*len(ad_end)/rate_ad_end))
        ad_end = signal.resample(ad_end, num)

    x_ad_start = np.arange(0, ad_start.size) / rate_ad_start
    x_ad_end = np.arange(0, ad_end.size) / rate_ad_end

    x_source = np.arange(0, source.size) / rate

    z_ad_start = signal.correlate(source, ad_start, mode='same')
    z_ad_end = signal.correlate(source, ad_end, mode='same')

    x_z_ad_start = np.arange(0, z_ad_start.size) / rate
    x_z_ad_end = np.arange(0, z_ad_end.size) / rate

    peaks_ad_start = signal.find_peaks(
    z_ad_start,
    height=3500000000000000,
    distance=50000
    )
    peaks_ad_end = signal.find_peaks(
        z_ad_end,
        height=10000000000000000,
        distance=50000
    )

    peaks_idxs_ad_start = peaks_ad_start[0]
    peaks_idxs_ad_end = peaks_ad_end[0]
    array = [];
    for i, peak_idx in enumerate(peaks_idxs_ad_start):
        start = (peak_idx-ad_start.size/2) / rate
        start_m, start_s = divmod(start, 60)
        print(f"peak {i}: start {start_m:.0f} minutes, {start_s:.0f} seconds")
        array.append([filename,'start',i,start])

    for i, peak_idx in enumerate(peaks_idxs_ad_end):
        start = (peak_idx-ad_end.size/2) / rate
        start_m, start_s = divmod(start, 60)
        print(f"peak {i}: start {start_m:.0f} minutes, {start_s:.0f} seconds")
        array.append([filename,'end',i,start])
    end = time.time()
    print(time.process_time() - t)

    with open('spotify.csv','a', newline='') as f:
        for line in array:
            writer = csv.writer(f)
            writer.writerow(line)

        
        
    
    
```
